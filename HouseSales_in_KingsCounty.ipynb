```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

This data set contains house sale prices
for King County, which includes Seattle, Washington. It includes homes sold between May
2014 and May 2015.

## 1. Data Exploration

```python
#read file
dfH=pd.read_csv('kc_house_data.csv')
dfH
```

```python
#Get an overview of the data:
dfH.describe()
```

```python
#Get an overview of the data:
print(dfH.shape)
print(dfH.dtypes)
print(dfH.isnull().values.sum())
dfH.info()
```

```python
#Change the data format
dfH['date']=dfH['date'].str[:8]
dfH['date']= pd.to_datetime(dfH['date'])
dfH.head()
```

## 2. Hypothesis and Data Visualizations

#### 1.The number of bedrooms affects the house price.

```python
# Explore the relationship between number of bedrooms and house price

sns.violinplot(x='bedrooms', y='price', data=dfH,size=20)
plt.title("Relationship between Bedrooms and House Price")
plt.show()
```

Insight:The house price has the most fluctuations when there are 5 to 6 bedrooms. 
For houses with 0 to 4 bedrooms, the prices are relatively consistent and tend to increase with the number of rooms. 
However, when the number of bedrooms exceeds 6, there is no longer a positive correlation between the average house price and the number of bedrooms.

#### 2.The total square footage of living area affects the house price.

```python
# Explore the relationship between square footage of living area and house price
sns.scatterplot(x='sqft_living', y='price', data=dfH,s=5)
plt.show()
```

```python
df_sqft = dfH.copy().groupby(['sqft_living'])['price'].mean()
fig, ax = plt.subplots(figsize=(15,4))
df_sqft.plot(kind='line', ax=ax)
ax.set_ylabel("price")
ax.set_xlabel("living area")
ax.set(title="Average price by sqft living")
```

#### 3.The location of the house affects the house price.

```python
# Calculate the average house price for each zip code
grouped = dfH.groupby('zipcode').mean()
grouped = grouped.reset_index()

# Plot the average house price by zip code
sns.barplot(x='zipcode', y='price', data=grouped,edgecolor='black')

# Add a title and labels to the plot
plt.title("Average House Price by Location")
plt.xlabel("Location (Zip Code)")
plt.ylabel("Average House Price (USD)")
plt.xticks(rotation=90,fontsize=5.5)

# Show the plot
plt.show()
#insight:the zipcode is 98039 has the highest average house price the 09112 followed behind.
```

#### 4. The average house price is related to the date the house was sold

```python
# Create a line plot to explore the relationship between date and average house price
dfH_date_mean=dfH.groupby(['date'])['price'].mean()

fig,axs = plt.subplots(figsize=(15,4))
dfH_date_mean.plot(kind='line',ax=axs)
axs.set_ylabel('Price')
axs.set_xlabel('Year')
axs.set_title('Average price by year')
```

#### 5.The average house price is affected by the number of bedrooms, total square footage of living area, and the location of the house.

```python
# Calculate the average house price for each combination of bedrooms, square footage of living area, and zip code
grouped = dfH.groupby(['bedrooms', 'sqft_living', 'zipcode']).mean()
grouped = grouped.reset_index()

# Explore the relationship between the average house price and the three factors
sns.pairplot(data=grouped, x_vars=['bedrooms', 'sqft_living', 'zipcode'], y_vars='price',height=4, aspect=1,plot_kws={"s": 7})
plt.show()
```

#### 6.The average house price is affected by the grade of the house, year the house was built, and renovation history.

```python
# Calculate the average house price for each combination of grade, year built, and year renovated
grouped = dfH.groupby(['grade', 'yr_built', 'yr_renovated']).mean()
grouped = grouped.reset_index()

# Explore the relationship between the average house price and the three factors
sns.pairplot(data=grouped, x_vars=['grade', 'yr_built', 'yr_renovated'], y_vars='price', height=4, aspect=0.8,plot_kws={"s": 7})
plt.show()
```

In 2014-10 there is a peak time for the house price.

#### 7.Explore Correlation Matrix

```python
#Explore Correlation Matrix
import seaborn as sns
#Compute the correlation matrix
corr = dfH.corr()

#Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, cmap="coolwarm", annot_kws={"size": 6})
```

Overall, housing prices are most related to the size and grade of the living area.The grade is related to the  living area and the sqft of above.
